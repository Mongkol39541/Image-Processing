```python
import torch
import cv2
from pathlib import Path
import numpy as np
```

```python
weights_path = '/content/yolov5/runs/train/ModelX/weights/best.pt'
model = torch.hub.load('ultralytics/yolov5', 'custom', path=weights_path)

def detect_objects(image_path, model):
    img = cv2.imread(image_path)
    results = model(img)
    boxes = results.xyxy[0][:, :4].cpu().numpy()
    classes = results.pred[0][:, -1].cpu().numpy()
    scores = results.pred[0][:, 4].cpu().numpy()
    boxes_n = []
    for boxe in boxes:
      boxe = [int(i) for i in boxe]
      boxes_n.append(boxe)
    classes = [int(classe) for classe in classes]
    scores = [round(score, 4) for score in scores]
    return boxes_n, classes, scores
```

```python
import pandas as pd
from tqdm import tqdm
```

```python
df = pd.read_csv('/content/sample_submission.csv')
df
```

```python
b = []
c= []
s = []

for name in tqdm(df['id']):
  image_path = '/content/test/{0}'.format(name)
  boxes, classes, scores = detect_objects(image_path, model)
  b.append(boxes)
  c.append(classes)
  s.append(scores)

df['boxes'] = b
df['labels'] = c
df['scores'] = s
df
```

```python
df.to_csv('/content/submission1.csv', index=False)
```
